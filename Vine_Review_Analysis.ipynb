```python
#Dependencies
import pandas as pd

# Import vine_table (major appliances)
vine_table_to_load = "vine_table.csv"

#Read vine_table
vine_df = pd.read_csv(vine_table_to_load)

#show df
vine_df
```

```python
# new df where total_votes is >= 20

TwentyPlusVotes_df = vine_df[vine_df["total_votes"] >= 20]

#sort by ascending to double check
#TwentyPlusVotes_df.sort_values(by="total_votes")
```

```python
# New df where helpful_votes/total_votes >= 50%

fiftyPlusPercentHelpful_df = TwentyPlusVotes_df[TwentyPlusVotes_df["helpful_votes"]/TwentyPlusVotes_df["total_votes"] >= 0.5]

#show df
#fiftyPlusPercentHelpful_df
```

```python
# New df that retrieves all the rows where a review was written as part of the Vine program (paid)

paid_vine_user_df = fiftyPlusPercentHelpful_df[fiftyPlusPercentHelpful_df["vine"] == "Y"]

paid_vine_user_df
```

```python
# New df that retrieves all the rows where a review was written by unpaid users

unpaid_vine_user_df = fiftyPlusPercentHelpful_df[fiftyPlusPercentHelpful_df["vine"] == "N"]

unpaid_vine_user_df
```

```python
# Total number of reviews - Paid

total_reviews_paid = paid_vine_user_df.count()["review_id"]
total_reviews_paid
```

```python
#total number of 5 stars - Paid

paid_vine_user_FiveStar = len(paid_vine_user_df[paid_vine_user_df["star_rating"] == 5])
paid_vine_user_FiveStar
```

```python
# % of 5 stars - Paid

paid_vine_user_PercentFiveStar = paid_vine_user_FiveStar / total_reviews_paid 
paid_vine_user_PercentFiveStar
```

```python
# Total number of reviews - Unpaid

total_reviews_unpaid = unpaid_vine_user_df.count()["review_id"]
total_reviews_unpaid
```

```python
#total number of 5 stars - Unpaid

unpaid_vine_user_FiveStar = len(unpaid_vine_user_df[unpaid_vine_user_df["star_rating"] == 5])
unpaid_vine_user_FiveStar
```

```python
# % of 5 stars - Unpaid

unpaid_vine_user_PercentFiveStar = unpaid_vine_user_FiveStar / total_reviews_unpaid 
unpaid_vine_user_PercentFiveStar
```
